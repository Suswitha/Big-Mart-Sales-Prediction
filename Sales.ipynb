```python
mba['Item_Type_Extracted']=mba['Item_Identifier'].apply(lambda x:x[0:2])
mba['Item_Type_Extracted'].value_counts()
```

```python
#Binning values

# Specify the boundaries of the bins
bins = [-np.inf, 1990, 2000, 2010]
# Bin labels
labels = ['1980-1990', '1990-2000', '2000-2010']

mba['Outlet_Establishment_Year']=pd.cut(mba['Outlet_Establishment_Year'], 
                                                  bins=bins, labels=labels)
mba['Outlet_Establishment_Year'].value_counts().astype('object')
```

```python
sns.barplot(data=mba, x='Outlet_Establishment_Year', y='Item_Outlet_Sales', hue="Outlet_Location_Type")
plt.legend(loc='best')
plt.show()
```
